Importing Required Python Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Reading Data from excel files into dataframes using pandas
and displaying first dataset

```python
df1=pd.read_excel("walmart1.xlsx")
df2=pd.read_excel("walmart2.xlsx")

df1
```

Displaying second dataset

```python
df2
```

Merging df1 and df2 using concat function of pandas.

```python
merge=[df1,df2]
walmart_data=pd.concat(merge)
```

Displaying merged dataset

```python
walmart_data
```

Finding out null values

```python
walmart_data.isnull()
```

Displaying sum of number of null values in the merged dataset

```python
walmart_data.isnull().sum()
```

Dropping rows with null values and displaying fresh dataset

```python
walmart_data=walmart_data.dropna()
walmart_data
```

Displaying columns with numeric datatype

```python
walmart_data.select_dtypes(include=['number'])
```

```python
walmart_data.to_excel("Walmart.xlsx")
```
