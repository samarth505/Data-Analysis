Importing required libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
```

Creating a dataframe

```python
data = {
'Subject': ['English', 'Hindi', 'Maths', 'Science', 'Social Science'],
'Marks': [90, 85, 78, 92, 88],
'Attendance': [95, 95, 96, 90, 94]
}

df = pd.DataFrame(data)

df
```

Assigning Variables to columns for plotting

```python
Subject=df['Subject']
Marks=df['Marks']
Attendance=df['Attendance']
```

Plotting a Line graph

```python
plt.plot(Subject,Marks)
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Line graph showing marks in each subject')
plt.show()
```

Plotting a histogram

```python
plt.hist(Attendance)
plt.ylabel('Frequency')
plt.xlabel('Attendance(in %)')
plt.title('Histogram showing attendance')
plt.show()
```

Plotting a bar graph

```python
plt.bar(Subject,Attendance)
plt.xlabel("Subject")
plt.ylabel('Attendance (in %)')
plt.title("Bar graph showing attendance in each subject")
plt.show()
```

Plotting a pie chart

```python
plt.pie(Marks,labels=Subject, autopct='%1.1f%%')
plt.title("Pie chart showing marks in each subject")
plt.show()
```

Plotting a scatter plot

```python
plt.scatter(Subject,Marks)
plt.xlabel("Subject")
plt.ylabel('Marks')
plt.title("Scatter Plot showing marks in each subject")
plt.show()
```
