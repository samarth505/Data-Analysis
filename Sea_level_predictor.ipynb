```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np
```

```python
#Data Extracted from given CSV

df=pd.read_csv('epa-sea-level.csv')
df
```

```python
#Selected Required Columns
Years=df['Year']
Sea_Level=df['CSIRO Adjusted Sea Level']
```

```python
#Scatter Plot for All

plt.xlabel("Years")
plt.ylabel('Sea level (in inches)')
plt.title("Sea level through the years")
plt.scatter(Years,Sea_Level)
```

```python
#Linear regression calculation
slope, intercept, r_value, p_value, std_err = linregress(Years, Sea_Level)

#Line of best fit for all data
x = np.array(Years)
y=slope*x+intercept

#Plot line of best fit for all data
plt.xlabel("Years")
plt.ylabel("Sea level (in inches)")
plt.title("Line of best fit")
plt.plot(x,y,label='Line of best fit (All data)')
plt.legend()

```

```python
#Filtering data for years since 2000
df_since_2000=df[df['Year']>=2000]
#Selected required columns for data since 2000
Years_since_2000=df_since_2000['Year']
SeaLevel_since_2000=df_since_2000['CSIRO Adjusted Sea Level']

#Linear regression calculation for data since 2000
slope_since_2000,intercept_since_2000,r_value_since_2000,p_value_since_2000,std_err_since_2000=linregress(Years_since_2000, SeaLevel_since_2000)


#Line of best fit for data since 2000
x_since_2000=Years_since_2000
y_since_2000 = slope_since_2000 * x_since_2000 + intercept_since_2000
#Plot line of best fit for data since 2000
plt.xlabel("Years")
plt.ylabel("Sea level (in inches)")
plt.title('Line of best fit')
plt.plot(x_since_2000, y_since_2000,label='Line of best fit(Data since 2000)')
#Plot line of best fit for all data
plt.xlabel("Years")
plt.ylabel("Sea level (in inches)")
plt.title("Line of best fit")
plt.plot(x,y,label='Line of best fit (All data)')
plt.legend()

#Prediction
prediction_year = 2050

prediction_value_all_data = slope * prediction_year + intercept
prediction_value_data_since_2000 = slope_since_2000 * prediction_year  + intercept_since_2000

#Plotting sea level prediction
plt.scatter(prediction_year, prediction_value_all_data,label='Prediction(All data)')
plt.scatter(prediction_year, prediction_value_data_since_2000,label='Prediction(Data since 2000)')



# Add labels and legend
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
plt.legend()


# Display the plot
plt.show()


```
